### Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

num_steps = 1000
stepsize = 0.01

num_samples = 10
x_data = np.array(range(num_samples))
y_data = np.array(range(num_samples) + 0.1*np.random.rand(num_samples))

W0, b0 =  10, 1

W, b = W0, b0
for i in range(num_steps):
    y_pred = np.dot(W,x_data) + b
    grad_w = (2.0/(num_samples))*sum((y_pred - y_data)*x_data)
    grad_b = (2.0/(num_samples))*sum((y_pred - y_data))
    
    W -=  stepsize*grad_w
    b -=  stepsize*grad_b
    
plt.plot(x_data,y_data)    
print W, b
```

### Neural nets

```python
class Net:
    def __init__(self, input_dim, output_dim):
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.hidden_dim = 2*input_dim
        self.net = self.make_net() 
        self.batchsize = 10
        self.lr = 0.01
        self.num_steps = 1000
        
    def make_net(self):
        self.W1 = np.random.rand(self.hidden_dim,self.input_dim, )
        self.b1 = np.random.rand(self.hidden_dim)
        
        self.W2 = np.random.rand(self.output_dim,self.hidden_dim)
        self.b2 = np.random.rand(self.output_dim)
        
        
    def sigmoid(self,x):
        return 1.0 / (1.0 + np.exp(-x))
    
    
    def grad_sigmoid(self,x):
        return np.exp(-x) / ( np.exp(-x) + 1.0 )**2 
        
        
    def feed_forward(self,a0):
        
        #Find activations
        z1 = np.dot(net.W1,a0.T) +  np.array([net.b1 for i in range(num_samples)]).T
        a1 = self.sigmoid(z1)    #this is a stack of vectors
        
        z2 = np.dot(net.W1,a1.T) +  np.array([net.b2 for i in range(num_samples)]).T
        a2 = self.sigmoid(z2)    # 
        
        return a2
    
    
    def find_gradients(self, x, y):
        """ Find gradients """
        
        a0 = x
        
        #Find activations
        z1 = np.dot(net.W1,a0.T) +  np.array([net.b1 for i in range(num_samples)]).T
        a1 = self.sigmoid(z1)    #this is a stack of vectors
        
        z2 = np.dot(net.W1,a1.T) +  np.array([net.b2 for i in range(num_samples)]).T
        a2 = self.sigmoid(z2)    # """
                
        #Find deltas
        delta2 = np.mean(a2 - y, axis = 0)   #this is for Mean Sq Error
        delta1 = np.dot(self.W2.T, delta2)*self.grad_sigmoid(z1)
        
        #Return grads
        grad_W2 = np.outer(a1, delta2)
        grad_W1 = np.outer(a0, delta1)
        
        grad_b2 = delta2
        grad_b1 = delta1
        
        return grad_W1, grad_W2, grad_b1, grad_b2
        
    
    def train(self, x, y):
        
        #Stochastic gradient descent
        num_samples = x.shape[1]
        for i in range(self.num_steps):
            indicies = np.random.choice(range(num_samples),self.batchsize)
            x_batch, y_batch = x[indicies], y[indicies]

            grad_W1, grad_W2, grad_b1, grad_b2 = self.find_gradients(self,x_batch,y_batch)

            self.W1 -= self.lr*grad_W1
            self.W2 -= self.lr*grad_W2

            self.b1 -= self.lr*grad_b1
            self.b2 -= self.lr*grad_b2
         
        
input_dim, output_dim = 3,2            
net = Net(input_dim, output_dim)

num_samples = 10
x_data = np.random.rand(num_samples, input_dim)
y_data = np.random.rand(num_samples, output_dim)

y_pred = net.feed_forward(x_data)
y_pred.shape
```

### Projected gradient descent

Gradient descent with a constraint

https://math.stackexchange.com/questions/571068/what-is-the-difference-between-projected-gradient-descent-and-ordinary-gradient

Require $\theta \in C$ -- some constraint, say all positive

So, 

\begin{equation}
   \theta_{n+1} = \theta_{n} - \alpha_n * \nabla C \\
   \theta_{n+1} = argmin_{x \in C} || \theta_{n+1} - x ||   \rightarrow \text{find theta closest to the constraint}
\end{equation}

### K means

Lyyods algorithm

1. Set centers randomly
2. Assign points to nearest center
3. Update center by mean of points within the cluster
4. Repeat till some convergence  (centroids don't change, clusters -- set of points assigned to certain x -- don't change)

Sensitive to outliers (since its takes means) -- K-medians get around this

1. Fails on non-convex data, and non densities


https://www.datascience.com/blog/k-means-clustering

```python
def dist(x,y):
    return sum((y - x)**2)


def Kmeans(data, K):
    
    #Initial centroids
    maxes= np.max(x_data, axis=0)
    mins = np.min(x_data, axis=0)
    centroids = [np.array([np.random.uniform(mins[i], maxes[i]) for i in range(len(maxes))]) for j in range(K)]

    num_steps = 100
    trial = 0
    while trial < num_steps:

        #Assign to clusters
        clusters = {i:[] for i  in range(K)}
        for i,point in enumerate(x_data):
            label = np.argmin([dist(point, centroid) for centroid in centroids])
            clusters[label].append(point)

        ##Update centroids
        for i in range(K):
            new_centroid = np.mean(np.array(clusters[i]),axis=0)
            centroids[i] = new_centroid

        #Check cond
        trial += 1
    return clusters, centroids
```

```python
from sklearn.datasets import load_iris


#Example dataset from sklearn
iris = load_iris()
x_data = iris.data[:, :2]        #only take the first few features
y_data = (iris.target != 0) * 1  #
plt.scatter(x_data[:,0],x_data[:,1], c=y_data, edgecolors='k', cmap=plt.cm.Paired)

K = 2
clusters, centroids = Kmeans(x_data,K)

for i in range(K):
    plt.scatter(centroids[i][0], centroids[i][1], marker = 'x')
```
