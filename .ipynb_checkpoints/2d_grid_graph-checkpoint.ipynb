### Time series

```python
def randomwalk_ncovertime_series(G,n,T):
    """
    Same as above, except now a taxi trip takes a radom walk,
    starting at 
    """

    nodes = [node for node in G.nodes()]
    Cs = []
    counts = {edge:0 for edge in G.edges()}       
    time = 0
    current_position = np.random.choice(nodes)
    num_unvisited_edges = np.sum(np.array(counts.values()) < n)
    Cs.append(num_unvisited_edges)
    while time < T:
        time += 1
        neighbours = G[current_position].keys()
        new_position = np.random.choice(neighbours)
        counts[(current_position,new_position)] += 1
        current_position = new_position
        num_unvisited_edges = np.sum(np.array(counts.values()) < n)
        Cs.append(num_unvisited_edges)
    return Cs



def randomdrive_ncovertime_series(G,n,T):
    """
    Same as above, except now a taxi trip takes a random walk,
    starting at 
    """

    nodes = [node for node in G.nodes()]
    Cs = []
    time = 0
    counts = {edge:0 for edge in G.edges()}       
    current_position = np.random.choice(nodes)
    num_unvisited_edges = np.sum(np.array(counts.values()) < n)
    Cs.append(num_unvisited_edges)
    while time < T:
        nodes_minus_origin = nodes[:current_position] + nodes[current_position+1:]
        destination = np.random.choice(nodes_minus_origin)
        path = nx.shortest_path(G,current_position, destination)
        for i in range(len(path)-1):
            edge = (path[i], path[i+1])
            counts[edge] += 1
            num_unvisited_edges = np.sum(np.array(counts.values()) < n)
            time += 1
            Cs.append(num_unvisited_edges)
        current_position = destination
    return Cs
```

```python
N,p = 10**2, 0.1
G = nx.gnp_random_graph(N,p)
G = list(nx.connected_component_subgraphs(G))[0]  #extract largest CC

n, T = 1, 10**4
t1 = time.time()
Cs_rw = randomwalk_ncovertime_series(G,n,T)
Cs_td = randomdrive_ncovertime_series(G,n,T)
t2 = time.time()
print 'took ' + str( (t2-t1)/60.0 ) + ' mins'
plt.plot(Cs_rw)
plt.plot(Cs_td)
plt.legend(['Random walk', 'Taxi drive'])
plt.xlabel('time',fontsize=16);plt.ylabel('Num unvisited edges')
```
